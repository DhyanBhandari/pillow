### TESTING THE Health-MODEL:

```python
#.................................................Health_prediction(TESTING)............................................#

import os
import time
import joblib
import warnings
import pandas as pd
warnings.filterwarnings(action = 'ignore')
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split


def Health_model(Health_value):    
    knn = joblib.load('Health_Model.pkl')
    x_new = Health_value
    l=[]
    for i in range(len(x_new)):
        x_new[i]=float(x_new[i])
    l.append(x_new[2])
    l.append(x_new[1])
    l.append(x_new[0])
    print(l)
    try:
        new_pred = knn.predict([l])
    except:
        print("error")
    print('new_pred',new_pred)
    if new_pred==[0.]:
        result = 'Normal'
        print(result)
    elif new_pred==[1.]:
        result = 'Moderate'
        print(result)
    else:
        result = 'Severe'
        print(result)
    f = open("output.txt","w")
    f.write(result)
    f.close()
    print('\n')


if __name__ == '__main__':
    TestData="Health_Test"
    while True:
        for(direcpath,direcnames,files) in os.walk(TestData):
            for file in files:
                if 'txt' in file:
                    try:
                        time.sleep(1)
                        filename= TestData +'/'+ file
                        print('Filename:',filename)
                        with open(filename, 'r') as myfile:
                            data = myfile.read()
                            testdata = data.split("#")
                            print(testdata)
                            clf = Health_model(testdata)
                        #os.remove(filename)
                    except:
                        pass
```
